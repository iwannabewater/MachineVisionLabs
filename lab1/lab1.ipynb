### Task 1: Balance in Brightness

**Objective**: Create a nighttime version of a daylight image, where the tree and its immediate surroundings are illuminated, but the background is dark. This will be achieved using intensity inversion and gamma correction.

### Step-by-Step Implementation

#### 1. Read the Input Image
First, read the input image using OpenCV.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the input image
input_image_path = './images/Task1.jpg'
image = cv2.imread(input_image_path)

# Display the original image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
plt.show()
```

#### 2. Convert to Grayscale
Convert the image to grayscale as the final output needs to be in grayscale.

```python
# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

#### 3. Apply Intensity Inversion
Perform intensity inversion by subtracting pixel values from 255.

```python
# Apply intensity inversion
inverted_image = 255 - gray_image

# Display the inverted image
plt.imshow(inverted_image, cmap='gray')
plt.title('Inverted Image')
plt.axis('off')
plt.show()
```

#### 4. Apply Gamma Correction
Use gamma correction to adjust the brightness of the image.

```python
# Apply gamma correction
gamma = 2.6
gamma_corrected_image = np.array(255 * (inverted_image / 255) ** gamma, dtype='uint8')

# Display the gamma corrected image
plt.imshow(gamma_corrected_image, cmap='gray')
plt.title('Gamma Corrected Image')
plt.axis('off')
plt.show()
```

#### 5. Apply Further Darkening to the Background
Create a mask to selectively darken the background while keeping the tree and its surroundings illuminated.

```python
# Create a mask for the tree area based on the original grayscale image
_, mask = cv2.threshold(gray_image, 80, 255, cv2.THRESH_BINARY_INV)

# Apply the mask to the gamma corrected image to darken the background more
background_darkened_image = gamma_corrected_image.copy()
background_darkened_image[mask == 255] = background_darkened_image[mask == 255] * 0.3

# Display the final nighttime effect image
plt.imshow(background_darkened_image, cmap='gray')
plt.title('Nighttime Effect Image')
plt.axis('off')
plt.show()
```

#### 6. Display Original and Final Images Side by Side
Display the original and final images side by side for comparison.

```python
# Display the original and final output image side by side
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].imshow(gray_image, cmap='gray')
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(background_darkened_image, cmap='gray')
axs[1].set_title('Nighttime Effect Image')
axs[1].axis('off')

plt.show()
```

### Conclusion
The above steps successfully convert the daylight image into a nighttime version where the tree and its immediate surroundings are illuminated against a dark background.

### Task 2: Noise Reduction in an Image

**Objective**: Reduce the noise in the provided image using various filtering methods and determine which method works best.

### Step-by-Step Implementation

#### 1. Read the Input Image
First, read the input image using OpenCV.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the input image
input_image_path = './images/Task2.jpg'
noisy_image = cv2.imread(input_image_path, cv2.IMREAD_GRAYSCALE)

# Display the original noisy image
plt.imshow(noisy_image, cmap='gray')
plt.title('Noisy Image')
plt.axis('off')
plt.show()
```

#### 2. Apply Mean (Uniform) Filtering
Use a mean filter to reduce the noise.

```python
# Apply mean (uniform) filtering
mean_filtered_image = cv2.blur(noisy_image, (5, 5))

# Display the mean filtered image
plt.imshow(mean_filtered_image, cmap='gray')
plt.title('Mean Filtered Image')
plt.axis('off')
plt.show()
```

#### 3. Apply Gaussian Filtering
Use a Gaussian filter to reduce the noise.

```python
# Apply Gaussian filtering
gaussian_filtered_image = cv2.GaussianBlur(noisy_image, (5, 5), 0)

# Display the Gaussian filtered image
plt.imshow(gaussian_filtered_image, cmap='gray')
plt.title('Gaussian Filtered Image')
plt.axis('off')
plt.show()
```

#### 4. Apply Median Filtering
Use a median filter to reduce the noise.

```python
# Apply median filtering
median_filtered_image = cv2.medianBlur(noisy_image, 5)

# Display the median filtered image
plt.imshow(median_filtered_image, cmap='gray')
plt.title('Median Filtered Image')
plt.axis('off')
plt.show()
```

#### 5. Compare the Results
Display the original and filtered images side by side for comparison.

```python
# Display the original and filtered images side by side
fig, axs = plt.subplots(1, 4, figsize=(20, 5))
axs[0].imshow(noisy_image, cmap='gray')
axs[0].set_title('Noisy Image')
axs[0].axis('off')

axs[1].imshow(mean_filtered_image, cmap='gray')
axs[1].set_title('Mean Filtered Image')
axs[1].axis('off')

axs[2].imshow(gaussian_filtered_image, cmap='gray')
axs[2].set_title('Gaussian Filtered Image')
axs[2].axis('off')

axs[3].imshow(median_filtered_image, cmap='gray')
axs[3].set_title('Median Filtered Image')
axs[3].axis('off')

plt.show()
```

### Conclusion
Based on the visual inspection, the median filtering method provides the best noise reduction while preserving the edges and details of the image. Median filtering is particularly effective at reducing salt-and-pepper noise, which is common in noisy images like the one provided.

### Response and Explanation

#### Filtering Methods Used
1. **Mean Filtering**: This method calculates the average of the pixel values in the neighborhood and assigns this average value to the central pixel. It is effective in reducing Gaussian noise but tends to blur the image, leading to loss of fine details.

2. **Gaussian Filtering**: This method uses a Gaussian kernel to weight the neighborhood pixels, giving more importance to the central pixels. It provides a better smoothing effect compared to mean filtering and is effective in reducing Gaussian noise. However, it can still cause some blurring.

3. **Median Filtering**: This method replaces the central pixel value with the median of the pixel values in the neighborhood. It is particularly effective in reducing salt-and-pepper noise while preserving edges and fine details.

#### Best Method: Median Filtering
After applying all three filtering methods, I observed that the median filtering method provides the best noise reduction while preserving the edges and details of the image. The underlying principle of median filtering is that it effectively removes outlier pixel values (such as those caused by salt-and-pepper noise) without blurring the image. This makes it an excellent choice for noisy images where edge preservation is crucial.

### Task 3: Image Sharpening

**Objective**: Sharpen the provided blurred image using different sharpening methods and determine which method works best.

### Step-by-Step Implementation

#### 1. Read the Input Image
First, read the input image using OpenCV.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
image_path = './images/Task3.jpg'
blurred_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Display the original blurred image
plt.imshow(blurred_image, cmap='gray')
plt.title('Original Blurred Image')
plt.axis('off')
plt.show()
```

#### 2. Apply Laplacian Filtering
Use the Laplacian filter to sharpen the image.

```python
# Apply a Laplacian filter with a kernel size of 5 to enhance edges
laplacian = cv2.Laplacian(blurred_image, cv2.CV_64F, ksize=5)
# Create a sharpened image by enhancing details with the Laplacian filter
sharpened_image_laplacian = cv2.convertScaleAbs(blurred_image - 0.7 * laplacian)

# Display the Laplacian sharpened image
plt.imshow(sharpened_image_laplacian, cmap='gray')
plt.title('Laplacian Sharpened Image')
plt.axis('off')
plt.show()
```

#### 3. Apply Unsharp Masking
Use the unsharp masking technique to sharpen the image.

```python
# Apply Gaussian blur as a part of the unsharp masking process
gaussian_blur = cv2.GaussianBlur(blurred_image, (3, 3), 2.0)
# Enhance the blurred image by adding weighted original and blurred images
unsharp_mask = cv2.addWeighted(blurred_image, 1.5, gaussian_blur, -0.5, 0)

# Display the unsharp masked image
plt.imshow(unsharp_mask, cmap='gray')
plt.title('Unsharp Masked Image')
plt.axis('off')
plt.show()
```

#### 4. Compare the Results
Display the original and sharpened images side by side for comparison.

```python
# Display the original and sharpened images side by side
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.title('Original Blurred Image')
plt.imshow(blurred_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Unsharp Masked Image')
plt.imshow(unsharp_mask, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Laplacian Sharpened Image')
plt.imshow(sharpened_image_laplacian, cmap='gray')
plt.axis('off')

# Show the plot
plt.show()
```

### Response and Explanation

#### Sharpening Methods Used
1. **Unsharp Masking**: This technique enhances edges by subtracting a blurred version of the image from the original image. It effectively sharpens the image by increasing contrast at the edges.
2. **Laplacian Filtering**: This technique uses the second derivative to detect edges and enhance them. It is a simple and effective way to sharpen images but can amplify noise.

#### Best Method: Laplacian Filtering
After applying both sharpening methods, I observed that Laplacian filtering provides a clearer sharpening effect. The underlying principle of Laplacian filtering is that it enhances edges by using the second derivative, which effectively highlights the edges and details in the image. This makes it an excellent choice for sharpening blurred images. But this is only the case for a specific single image, and the results of further exploration may be different.
